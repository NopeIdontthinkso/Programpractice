# Data Visualization with [Matplotlib](https://matplotlib.org/)


```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x_data = [0, 1, 2, 3]
height_data = [100, 200, 150, 50]
plot = plt.bar(x_data, height_data, bottom=0, color='#ff00ff') 
```

```python
height_data = [100, 200, 150, 50]
x = []
for i in range(len(height_data)):
    x.append(i)
plot = plt.bar(x, height_data, bottom=0, color='#ff00ff') 
```

```python
height_data = [100, 200, 150, 50]
plot = plt.bar(range(len(height_data)), height_data, bottom=0, color='#ff00ff') 
```

```python
height_data = [100, 200, 150, 50, 900 ,50 ,50]
x_labels = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
x_positions = range(len(x_labels))

plot = plt.bar(x_positions, height_data, bottom=0, color='#ff00ff')
plt.xticks(x_positions, x_labels)
plt.title('Apples', fontdict = {'fontsize' : 40})
```

```python
def create_bar(title, heights, x_labels, color, fontsize):
    x_positions = range(len(x_labels))
    height_data = heights
    
    plt.bar(x_positions, height_data, bottom=0, color=color)
    plt.xticks(x_positions, x_labels)
    plt.title(title, fontdict = {'fontsize' : fontsize})
    
def create_plot(title, heights, x_labels, color, fontsize):
    x_positions = range(len(x_labels))
    height_data = heights
    
    plt.plot(x_positions, height_data, color=color)
    plt.xticks(x_positions, x_labels)
    plt.title(title, fontdict = {'fontsize' : fontsize})
```

```python
height_data = [100, 200, 150, 50, 900 ,50 ,50]
x_labels = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
color='#ff00ff'
create_bar('apple', height_data, x_labels, color, 50)
```

```python
create_plot('apple', height_data, x_labels, color, 50)
```
